## Summary 

##### The data I chose to analyze is called 'Worlds Happiness Report'.  All data is gathered through the Gallup World Poll survery. The Gallup World Poll "tracks the most important issues worldwide, such as food access, employment, leadership performance, and well-being" (Gallup.com, 2019). The numbers in each column represent scores based on answers provided to questions asked in the poll. If you add all numbers horizontally in one line per each country, the sum will be the same as the 'Score'. The higher the score, the more happier the country is. 








##### I chose this data to work with because I wanted to examine which factors contribute to the happiness score the most. I worked with data from 2016-2019. In conclusion, this was by far the cleanest data set that I have worked with.

## Merging data together 

```python
#2019
df = df[['Country or region','Score','GDP per capita',
         'Healthy life expectancy','Freedom to make life choices','Generosity','Perceptions of corruption']]

#2018
df2 = df2[['Country or region','Score','GDP per capita',
         'Healthy life expectancy','Freedom to make life choices','Generosity','Perceptions of corruption']]

#2016                       
df4 = df4[['Country','Happiness Score','Economy (GDP per Capita)','Family','Health (Life Expectancy)',
'Trust (Government Corruption)','Generosity']]


# 2017 
df3 = df3[['Country','Happiness.Score','Economy..GDP.per.Capita.','Family','Health..Life.Expectancy.',
           'Generosity','Trust..Government.Corruption.' ]]


new_names = ['Country', 'Happiness Score', 'Economy (GDP per Capita)', 'Family',
                       'Health (Life Expectancy)',  'Generosity', 'Trust (Government Corruption)', 
                       ]

df.columns = new_names
df2.columns = new_names
df3.columns = new_names
df4.columns = new_names

df['Year'] = 2019
df2['Year'] = 2018
df3['Year'] = 2017
df4['Year'] = 2016

data = pd.concat([df, df2,df3,df4], axis=0)
data.head(5)
```

### Let's see which factor contributes to the Happiness Score the least and the most by Average 

```python
data={
    'Year':['2019', '2018', '2017', '2016'],
    'Economy (GDP per Capita)' : [0.905147,0.891449,0.984718,0.953880],
    'Happiness Score':[5.407096,5.375917,5.354019, 5.382185],
    'Health (Life Expectancy)': [0.725244,0.597346,0.551341,0.557619]
}
d=pd.DataFrame(data)
d
```

```python
ax = d.plot(y="Happiness Score", x = 'Year' , kind="bar",color='C4')
d.plot(y='Economy (GDP per Capita)', x = 'Year', kind="bar", ax=ax, color="C2")
d.plot(y='Health (Life Expectancy)',x = 'Year', kind="bar", ax=ax, color="C3")
plt.show()
```

##### Looking at the average by year, looks like the Life Expectancy Factor contributes more to the Happiness Score rather than the GPD Per Capita Factor. 

### Let's see the corrolation between these two factors

```python
plt.figure(figsize=(10,10))
plt.scatter(data['Economy (GDP per Capita)'], data['Health (Life Expectancy)'], s=(data['Happiness Score']**3), alpha=0.3)
plt.grid(True)

plt.xlabel("Economy")
plt.ylabel("Life Expectancy")

plt.suptitle("Corrolation between Life Expectancy and GPA Facotor", fontsize=15)

plt.show()
```

##### We can see there is a highly saturated area between 1.0 and 1.5 for GPA Facor. We can also see that the higher the economic GDP Factor, the higher the life expectancy. This conculdes that there is a strong corrolation betweeen these two factors. As the x-axis increases, so does the y-axis.
